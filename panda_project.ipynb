```python
# 1) Student Performance Analysis

# Dataset: Students marks dataset (subject-wise marks).

# Tasks to Perform:

# Load the data using pandas.

# Find total marks and percentage for each student.

# Identify highest scorer per subject.

# Plot bar/line chart for marks comparison using matplotlib.

# Skills Covered: Data cleaning, basic operations, sorting, filtering.

import pandas as pd
df = pd.read_csv("student_marks.csv")
df["Total"] = df["English"] + df["Maths"] + df["Science"]
df["Percentage"] = round((df["Total"] / 300) * 100,2)
print(df)
# print(df.set_index("Student"))
# print(df["English"].argmax())
# print(df["Maths"].argmax())
# print(df["English"].idxmax())
print()
print("the topper of english is",df.loc[df["English"].argmax(), "Student"],"with",df["English"].max(),"marks")
print("the topper of maths is",df.loc[df["Maths"].argmax(),"Student"],"with",df["Maths"].max(),"marks")
print("the topper of science is",df.loc[df["Science"].argmax(),"Student"],"with",df["Science"].max(),"marks")
```

```python
import matplotlib.pyplot as plt
students=("Arjun","Meena","Ravi","Sita","Kiran","Lakshmi","Rahul","Priya")
english=(78,88,67,90,72,85,60,95)
maths=(82,92,70,95,68,80,75,98)
science=(80,89,72,92,70,83,65,97)
plt.plot(students,english,marker="*",label="english",color="orange")
plt.plot(students,maths,marker="d",label="maths",color="y")
plt.plot(students,science,marker="p",label="science",color="green")
plt.xlabel("marks",color="g")
plt.ylabel("students",color="g")
plt.title("MARKS REPORT",color="r")
plt.legend()
plt.grid(True,linestyle="--",color="blue")
plt.show()
```

```python
# 2) Sales Data Analysis

# Dataset: Supermarket sales data (Date, Product, Quantity, Price).

# Tasks to Perform:

# Calculate total revenue = Quantity  Price.

# Group by product to find top-selling products.

# Group by month to find seasonal trends.

# Plot sales trend.

# Skills Covered: groupby, aggregation, date handling.


import pandas as pd
df=pd.read_csv("supermarket_sales.csv")
print(df)
```

```python

import pandas as pd
df=pd.read_csv("supermarket_sales.csv")
df["total_revenue"]=df["Quantity"]*df["Price"]
print(df)
print("-------------------------------------------------------------------------------")
print("the top of the selling products and total revenue")
print(df.groupby("Product")["total_revenue"].max().sort_values(ascending=False))
df["Month"] = pd.to_datetime(df["Date"]).dt.to_period("D")
print(df)
print("=======================================================================")
print(df.groupby("Month")["Quantity"].sum() )
# print(df[df["Date"] == "2024-02-21"].groupby("Month")["Quantity"].count())


```

```python
from matplotlib.lines import lineStyles
# Plot sales trend.
import matplotlib.pyplot as plt
teends=("2024-01-jan","2024-02-feb","2024-03-mar","2024-04-apr")
revenue=(23364,20657,24112,9948)
plt.plot(teends,revenue,marker="*",label="revenue")
plt.xlabel("months wise",color="g")
plt.ylabel("revenue amount",color="g")
plt.title("MONTHLY SALES",color="r")
plt.grid(True,linestyle="--")
plt.legend()
plt.show()

```

```python
# 3) Movie Ratings Analysis

# Dataset: IMDB or TMDB movie dataset (Rating, genre, year).

# Tasks to Perform:

# Find highest & lowest rated movies.

# Count movies released each year.

# Analyze which genre has highest average rating.

# Visualize Genre vs Average Rating.

# Skills Covered: value_counts, filtering, grouping, visualization.

import pandas as pd
df=pd.read_csv("movies.csv")
print(df)
```

```python
import pandas as pd
df=pd.read_csv("movies.csv")
print("the highest rated ",df.loc[df["Rating"].argmax()])   # Find highest
print("###"*50)
print("the lowest rated ",df.loc[df["Rating"].argmin()])     #lowest rated movies.
print("###"*50)
print(df["Year"].value_counts().sort_index())                 # Count movies released each year.

# Analyze which genre has highest average rating.
print("-----------------------------------------------------------------")
print("the most highest gener rating=",df.loc[df["Rating"].argmax(),"Genre"])

print("##"*75)

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("movies.csv")
a=df.groupby("Genre")["Rating"].mean()
print(a)
plt.figure(figsize=(10,2))
plt.bar(a.index,a.values)
plt.show()

```

```python
## Calculate average monthly temperature.
import pandas as pd
i=("jan","feb","mar","apr","may","jun","jul","sep","oct","nov","dec")
df = pd.read_csv("weather_dataset.csv")
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
avg_monthly_temp =round(df.groupby("Month")["Temperature"].mean(),2)
print(avg_monthly_temp)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
i=("jan","feb","mar","apr","may","jun","jul","sep","oct","nov","dec")
df = pd.read_csv("weather_dataset.csv")
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
avg_monthly_temp =round(df.groupby("Month")["Temperature"].max(),2)

plt.plot(avg_monthly_temp.index,avg_monthly_temp.values,marker="*")
plt.xlabel("months")
plt.ylabel("temperature")
plt.title("yearly temperature")
plt.grid(True,linestyle="--")
plt.show()
```
